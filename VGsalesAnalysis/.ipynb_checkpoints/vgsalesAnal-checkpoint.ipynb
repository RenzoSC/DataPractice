# Which year had the highest sales worldwide?

```python
data_year = data.groupby(by=['Year'])['Global_Sales'].sum()
#agrupa en base al año y se fija en GLOBAL SALES, luego hace la suma de las ventas en cada año
data_year=pd.DataFrame(data_year.reset_index())
#devuelve el index a default y lo convierte en un DATAFRAME para luego poder ser usado en barplot
data_year
```

```python
plt.figure(figsize=(15, 10))
sns.barplot( data=data_year,x='Year', y="Global_Sales")
plt.xticks(rotation=90)
```

# 5. Which genre game has been released the most in a single year?

```python
year_max_df = data.groupby(['Year', 'Genre']).size().reset_index(name='count')
year_max_idx = year_max_df.groupby(['Year'])['count'].transform(max) == year_max_df['count']
year_max_genre = year_max_df[year_max_idx].reset_index(drop=True)
year_max_genre = year_max_genre.drop_duplicates(subset=["Year", "count"], keep='last').reset_index(drop=True)
genre = year_max_genre['Genre'].values
genre
```

```python
year_max_df = pd.DataFrame(year_max_df)
year_max_df
```

```python
year_max_idx = pd.DataFrame(year_max_idx)
year_max_idx
```

```python
year_max_genre = pd.DataFrame(year_max_genre)
year_max_genre
```

```python
x = pd.DataFrame(year_max_df.groupby(['Year'])['count'].transform(max))
x
```

```python
plt.figure(figsize=(30, 15))
g = sns.barplot(x='Year', y='count', data=year_max_genre)
index=0
for value in year_max_genre['count'].values:
#     print(asd)
    g.text(index, value +5 , str(genre[index] + '----' +str(value)), color='#000', size=22, rotation= 90, ha="center")
    index += 1
plt.xticks(rotation=90, size=22)
plt.yticks(size=22)
plt.show()
```
